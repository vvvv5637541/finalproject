```python
def fiveday(s):
    time.sleep(3)
    # 這是抓取歷史資料
    stock_6207 = twstock.Stock(s)

    price_6207 = stock_6207.price[-5:]       # 近五日之收盤價
    high_6207 = stock_6207.high[-5:]         # 近五日之盤中高點
    low_6207 = stock_6207.low[-5:]           # 近五日之盤中低點
    date_6207 = stock_6207.date[-5:]         # 近五日的日期

    #print('price_6207--->', type(price_6207))
    #print('high_6207--->', high_6207)
    #print('low_6207--->', low_6207)
    #print('date_6207--->', date_6207)
    total = 'price--->' + str(price_6207) + ' high--->' + str(high_6207) + ' low--->' + str(low_6207)
    return total
```

```python
import twstock

def realtime(s):
    time.sleep(3)
    stock_real = twstock.realtime.get(s)
    # 抓取多個股票的方式 twstock.realtime.get(['2330', '2337', '2409'])
    return str(stock_real)
```
